## Data Preprocessing

```python
df['Reviews'].unique()
```

```python
df['Reviews'].str.isnumeric().sum()
```

```python
df[~df['Reviews'].str.isnumeric()]
```

```python
df_copy = df.copy()
```

```python
df_copy=df_copy.drop(df_copy.index[10472])
```

```python
df_copy[~df_copy['Reviews'].str.isnumeric()]
```

```python
#Converting reviews to int 
df_copy['Reviews'] = df_copy['Reviews'].astype(int)
df_copy.info()
```

```python
df_copy['Size'].unique()
```

```python
df_copy['Size']=df_copy['Size'].str.replace('M','000')
df_copy['Size']=df_copy['Size'].str.replace('k','')
df_copy['Size']=df_copy['Size'].replace('Varies with device',np.nan)
df_copy['Size']=df_copy['Size'].astype(float)
```

```python
df_copy['Size']
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        df_copy[cols]=df_copy[cols].str.replace(item,'')
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Installs']=df_copy['Installs'].astype('int')
df_copy['Price']=df_copy['Price'].astype('float')
```

```python
df_copy.info()
```

```python
## Handlling Last update feature
df_copy['Last Updated'].unique()
```

```python
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
df_copy['Day']=df_copy['Last Updated'].dt.day
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['Year']=df_copy['Last Updated'].dt.year
```

```python
df_copy.info()
```

```python
df_copy.head()
```

```python
df_copy.to_csv('data/google_cleaned.csv')
```

## EDA (Exploratory Data Analysis)

```python
clean_data = df_copy
```

- The dataset contains multiple instances of identical records, indicating the presence of duplicate data 

```python
clean_data = clean_data.drop_duplicates(subset=['App'],keep='first')
```

```python
clean_data.shape
```

* Features of Data

```python
num_categories = [feature for feature in clean_data.columns if clean_data[feature].dtype != 'O']
cat_categories = [feature for feature in clean_data.columns if clean_data[feature].dtype == 'O']
```

```python
print(f'Numerical Feature ({len(num_categories)} features):- {num_categories}')
print(f'Categorical Feature ({len(cat_categories)} features):- {cat_categories}')
```

```python
# Count of data on categorical columns
for col in cat_categories:
    print(clean_data[cols].value_counts(normalize=True)*100)
```

```python
#Viusalization
plt.figure(figsize=(15,15))
plt.title('Univarite Analysis',fontsize=20,fontweight='bold',alpha=0.8,y=1.)
for i in range(0,len(num_categories)):
    plt.subplot(6,3,i+1)
    sns.kdeplot(x=clean_data[num_categories[i]],fill=True,color='g')
    plt.xlabel(num_categories[i])
    plt.tight_layout()
```

1. Which Category has the largest number of installations

2. What are the Top 5 most installed apps in Each popular category
3. How many app are there on Google Play store with 5 rating ?


```python
cat_installs = df_copy.groupby(['Category'])['Installs'].sum().sort_values(ascending = False).reset_index()
cat_installs.Installs = cat_installs.Installs/1000000000# converting into billions
df2 = cat_installs.head(10)
plt.figure(figsize = (14,10))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(x = 'Installs' , y = 'Category' , data = df2 ,palette='hls')
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('')
ax.set_title("Most Popular Categories in Play Store", size = 20)
```

- "GAME" even though are not the top category of apps but have the highest number of installations



```python
most_installed= df_copy.groupby(['Category' ,'App'])['Installs'].sum().reset_index()
most_installed = most_installed.sort_values('Installs', ascending = False)
apps = ['GAME', 'COMMUNICATION', 'PRODUCTIVITY', 'SOCIAL' ]
sns.set_style("darkgrid")

plt.figure(figsize=(40,30))

for i,app in enumerate(apps):
    df2 = most_installed[most_installed.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data= df3,x='Installs' ,y='App',hue='Installs',legend=False,palette='hls')
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size = 20)
    
plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```

Observation : 

    - The most played game is "Subway Surfers"
    - The most used Social media app is "Instagram"
    - The most used (Productivity/Office) App is "GoogleDrive"

```python
rating = df_copy.groupby(['Category','Installs', 'App'])['Rating'].sum().sort_values(ascending = False).reset_index()

toprating_apps = rating[rating.Rating == 5.0]
print("Number of 5 rated apps",toprating_apps.shape[0])
toprating_apps.head(1)
```

Result 
- There are about 269 five star rated apps in the Google Play Store

- Top most is the "CN Resident" from "Family" category
